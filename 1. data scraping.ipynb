```python
import os
import pandas as pd

# Path ke folder "tweets-data"
folder_path = '/content/tweets-data'

# Inisialisasi DataFrame kosong untuk menyimpan hasil penggabungan
merged_df = pd.DataFrame()

# Iterate through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):  # Hanya proses file CSV
        file_path = os.path.join(folder_path, filename)

        # Check if the file is empty before attempting to read it
        if os.stat(file_path).st_size == 0:
            print(f"Skipping empty file: {filename}")
            continue  # Skip to the next file

        # Baca file CSV dan tambahkan ke DataFrame
        try:
            df = pd.read_csv(file_path, delimiter=";")
        except pd.errors.EmptyDataError:
            print(f"Skipping file with no columns: {filename}")
            continue

        # Tambahkan kolom baru yang berisi asal file CSV
        df['source_file'] = filename
        merged_df = pd.concat([merged_df, df], ignore_index=True)

# Simpan DataFrame hasil penggabungan menjadi satu file CSV
merged_file_path = '/content/merged_tweets.csv'
merged_df.to_csv(merged_file_path, index=False)

print("All CSV files have been merged into one with source column:", merged_file_path)
```

```python
# Read the CSV file into a pandas DataFrame
df_filtered = pd.read_csv("merged_tweets_all.csv", delimiter=";")
df_filtered
```

```python
# Calculate unique counts for each column
unique_counts = df_filtered.nunique()

# Print the unique counts
unique_counts
```

```python
import re

def extract_bushfire_status(text):
    # Define regex patterns
    pattern1 = r"Bushfire (.*?) for"
    pattern2 = r"Bushfire (.*?) -"

    # Try matching with pattern1
    match1 = re.search(pattern1, text)
    if match1:
        return match1.group(1)

    # Try matching with pattern2
    match2 = re.search(pattern2, text)
    if match2:
        return match2.group(1)

    # Return None if no match found
    return None

def split_full_text(text):
    """
    Splits full_text column into 5 new columns: date, time, status, location, and area.
    """
    # Define regular expressions for extracting information
    time_pattern = r'\d{2}:\d{2} [AP]M'
    location_pattern = r'of (.*?) in'
    area_pattern = r'in the (.*): '

    time_match = re.search(time_pattern, text)
    status_match = extract_bushfire_status(text)
    location_match = re.search(location_pattern, text)
    area_match = re.search(area_pattern, text)

    result = {
        "time": time_match.group() if time_match else None,
        "status": status_match if status_match else None,
        "location": location_match.group(1) if location_match else None,
        "area": area_match.group(1).strip() if area_match else None  # Remove trailing space for area
    }

    return result


# df_filtered = pd.read_csv("merged_tweets.csv")
for index, row in df_filtered.iterrows():
  # Extract information using split_full_text
  info = split_full_text(row["full_text"])

  # Assign extracted values directly to df_filtered using .loc
  df_filtered.loc[index, "time"] = info.get("time")
  df_filtered.loc[index, "status"] = info.get("status")
  df_filtered.loc[index, "location"] = info.get("location")
  df_filtered.loc[index, "area"] = info.get("area")

# Drop full_text
df_filtered.drop(columns=["full_text"], inplace=True)
```

```python
clean_df = df_filtered[['created_at', 'suburb', 'time', 'status']].copy()
clean_df
```

```python
clean_df['Date'] = pd.to_datetime(clean_df['created_at']).dt.strftime('%Y/%m/%d')
```

```python
# ubah nilai status menjadi huruf kapital
clean_df['status'] = clean_df['status'].str.upper()
clean_df
```

```python
clean_df = (
    clean_df
    .sort_values(by=['Date', 'suburb', 'time'])  # Pastikan urut berdasarkan waktu
    .groupby(['Date', 'suburb', 'status'])  # Kelompokkan berdasarkan Date, suburb, dan status
    .first()  # Ambil baris pertama dalam setiap kelompok
    .reset_index()  # Reset index setelah groupby
)
clean_df
```

```python
# jumlah status per bulan
clean_df['created_at'] = pd.to_datetime(clean_df['created_at'])
monthly_status_counts = clean_df.groupby(clean_df['created_at'].dt.to_period('M'))['status'].value_counts().unstack(fill_value=0)
monthly_status_counts
```

```python
# prompt: buat grafik yang menunjukkan frekuensi terjadinya per status, dan ubah grafik sebanyak jumlah subrub

import pandas as pd
import matplotlib.pyplot as plt

# # Load the cleaned data
# clean_df = pd.read_csv("cleanbushfire-juljan.csv")

# # Convert 'created_at' to datetime and extract month
# clean_df['created_at'] = pd.to_datetime(clean_df['created_at'])
clean_df['month'] = clean_df['created_at'].dt.month_name()

# # Convert status to uppercase
# clean_df['status'] = clean_df['status'].str.upper()

# Group by month and status and count the occurrences
monthly_status_counts = clean_df.groupby(['month', 'status']).size().unstack(fill_value=0)

# Plotting
monthly_status_counts.plot(kind='bar', figsize=(12, 6))
plt.title('Frequency of Bushfire Status by Month')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Status')
plt.tight_layout()
plt.show()


# Get unique suburb names
suburbs = clean_df['suburb'].unique()

# Loop through each suburb and plot
for suburb in suburbs:
  # Filter data for the current suburb
  suburb_data = clean_df[clean_df['suburb'] == suburb]

  # Group by month and status and count the occurrences
  monthly_status_counts = suburb_data.groupby(['month', 'status']).size().unstack(fill_value=0)

  # Plotting
  monthly_status_counts.plot(kind='bar', figsize=(12, 6))
  plt.title(f'Frequency of Bushfire Status by Month for {suburb}')
  plt.xlabel('Month')
  plt.ylabel('Frequency')
  plt.xticks(rotation=45, ha='right')
  plt.legend(title='Status')
  plt.tight_layout()
  plt.show()
```
