## Dataset burnt

```python
# go to folder data dec to feb
folder_path = 'DATA'
# burnt, danger, warning, advice data
burnt_path = os.path.join(folder_path, 'Burnt Areas.csv')
burnt = pd.read_csv(burnt_path, sep=';')# on_bad_lines='skip') # on_bad_lines='skip' will skip lines that cause errors.
burnt['created_at'] = pd.to_datetime(burnt['created_at']).dt.strftime('%Y/%m/%d')
burnt
```

```python
# jumlah status per bulan
burnt['created_at'] = pd.to_datetime(burnt['created_at'])
monthly_status_counts = burnt.groupby(burnt['created_at'].dt.to_period('M'))['status'].value_counts().unstack(fill_value=0)
monthly_status_counts
```

```python
burnt2_path="/DATA/cleanbushfire-martojan.csv"
burnt2 = pd.read_csv(burnt2_path, delimiter=';')# on_bad_lines='skip') # on_bad_lines='skip' will skip lines that cause errors.
burnt2['created_at'] = pd.to_datetime(burnt2['created_at']).dt.strftime('%Y/%m/%d')
burnt2
```

```python
burnt2 = burnt2[['created_at', 'time','suburb', 'status']]
burnt2
```

```python
# jumlah status per bulan
burnt2['created_at'] = pd.to_datetime(burnt2['created_at'])
monthly_status_counts = burnt2.groupby(burnt2['created_at'].dt.to_period('M'))['status'].value_counts().unstack(fill_value=0)
monthly_status_counts
```

```python
burnt = burnt[['created_at', 'time','suburb', 'status']]
burnt
```

```python
# Filter data for December, January and February
burnt_jan_feb = burnt[(burnt['created_at'].dt.month == 12)|(burnt['created_at'].dt.month == 1) | (burnt['created_at'].dt.month == 2)]

# Concatenate the filtered data with burnt2
combined = pd.concat([burnt_jan_feb, burnt2], ignore_index=True)
combined
```

```python
# Check for duplicates based on all columns
combined['suburb'] = combined['suburb'].str.strip()
combined['status'] = combined['status'].str.strip()
duplicates = combined[combined.duplicated(keep=False)]
combined.drop_duplicates(inplace=True)
```

```python
clean_df = (
    combined
    .sort_values(by=['created_at', 'suburb', 'time'])  # Pastikan urut berdasarkan waktu
    .groupby(['created_at', 'suburb', 'status'])  # Kelompokkan berdasarkan Date, suburb, dan status
    .first()  # Ambil baris pertama dalam setiap kelompok
    .reset_index()  # Reset index setelah groupby
)
clean_df
```

```python
clean_df = clean_df.rename(columns={'created_at': 'Date'})
clean_df['Date'] = pd.to_datetime(clean_df['Date'], errors='coerce')
clean_df['time'] = pd.to_datetime(clean_df['time'], errors='coerce').dt.hour
clean_df.loc[clean_df['time'] == 0, 'time'] = 24
clean_df
```

## Dataset weather

```python
from google.colab import drive
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
```

```python
# Define the folder path containing the CSV files
folder_path = os.path.join('DATA', 'Weather lengkap')

# Use glob to find all CSV files in the folder
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Create an empty list to store dataframes
df_list = []

# Loop through all CSV files
for filename in all_files:
    try:
        # Read each CSV file into a dataframe
        df = pd.read_csv(filename, sep=';', encoding='ISO-8859-1', on_bad_lines='skip')
        df_list.append(df)
        df['station'] = os.path.basename(filename)
    except pd.errors.ParserError:
        print(f"Error parsing file: {filename}. Skipping this file.")
    except Exception as e:  # Catch other potential errors during file reading
        print(f"An error occurred while processing {filename}: {e}")


# Concatenate all dataframes in the list
if df_list:  # Check if df_list is not empty
    combined_df = pd.concat(df_list, ignore_index=True)

    # Now you can work with combined_df, e.g., export to CSV
    combined_df.to_csv('combined_data.csv', index=False)
    print("Combined data saved to combined_data.csv")
else:
    print("No valid CSV files found in the specified directory.")
```

## Merged Dataset

```python
# data
file_path = os.path.join(folder_path, 'combined_data fixed.csv')
weather = pd.read_csv(file_path, sep=';')# on_bad_lines='skip') # on_bad_lines='skip' will skip lines that cause errors.
weather['Date'] = pd.to_datetime(weather['Date'], format='%d/%m/%Y').dt.strftime('%Y/%m/%d')
weather['Time of maximum wind gust'] = pd.to_datetime(weather['Time of maximum wind gust'], errors='coerce').dt.hour
weather.loc[weather['Time of maximum wind gust'] == 0, 'Time of maximum wind gust'] = 24
weather
```

```python
# Rename columns
new_column_names = {
    'Minimum temperature (Â°C)': 'Minimum Temperature',
    'Maximum temperature (Â°C)': 'Maximum Temperature',
    'Rainfall (mm)': 'Rainfall',
    'Evaporation (mm)': 'Evaporation',
    'Sunshine (hours)': 'Sunshine',
    'Direction of maximum wind gust ': 'Direction Max Wind Gust',
    'Speed of maximum wind gust (km/h)': 'Speed Max Wind Gust',
    'Time of maximum wind gust': 'Time Max Wind Gust'
}
weather = weather.rename(columns=new_column_names)

# Display or further process the combined DataFrame
weather
```

```python
weather.info()
```

```python
clean_df['Date'] = pd.to_datetime(clean_df['Date']).dt.strftime('%Y/%m/%d')
# Menambahkan kolom kosong kedalam dataframe 'burnt'
for col_name in ['Minimum Temperature', 'Maximum Temperature', 'Rainfall', 'Evaporation', 'Sunshine', 'Direction Max Wind Gust', 'Speed Max Wind Gust', 'Time Max Wind Gust']:
    clean_df[col_name] = '' # None or '' for empty string
clean_df
```

```python
clean_df.info()
```

```python
# Iterate through the 'clean_df' DataFrame
for index, row in clean_df.iterrows():
    date = row['Date']
    station = row['station']
    matching_rows = weather[(weather['Date'] == date) & (weather['station'] == station)]

    # If a match is found, update the corresponding columns in 'clean_df'
    if not matching_rows.empty:
        for col_name in ['Minimum Temperature', 'Maximum Temperature', 'Rainfall', 'Evaporation', 'Sunshine', 'Direction Max Wind Gust', 'Speed Max Wind Gust', 'Time Max Wind Gust']:
            clean_df.loc[index, col_name] = matching_rows.iloc[0][col_name]

clean_df
```

```python
clean_df.info()
```

```python
df= clean_df.copy()
```

```python
# Pastikan kolom numerik dikonversi ke tipe data float (karena masih dalam bentuk object)
numeric_columns = [
    "Minimum Temperature", "Maximum Temperature", "Rainfall",
    "Evaporation", "Sunshine", "Speed Max Wind Gust", "Time Max Wind Gust"
]

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Jika masih ada nilai null (karena semua dalam grup adalah NaN), isi dengan median keseluruhan kolom
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

df
```

```python
import numpy as np
# Ganti nilai kosong tersembunyi menjadi NaN
df.replace(["", " ", "NULL", "-"], np.nan, inplace=True)
```
